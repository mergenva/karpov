**Решение**

````python
# Объявляем функцию read_tasks, которая будет читать текстовый файл и записывать задачи в список
def read_tasks(text_file):
    # Создаем пустой список, где будут храниться преобразованные строки (задачи)
    tasks_list = []
    # Открываем файл
    with open(file=text_file, mode='r', encoding='utf-8') as f:
        # Обращаемся к каждой строке в файле
        for line in f:
            # Очищаем строку от пробелов по краям и добавляем очищенную строку в tasks_list
            strip_line = line.strip()
            tasks_list.append(strip_line)
    # Возвращаем список с задачами
    return tasks_list
````

````python
# Передаём в функцию переменную с названием файла и вызываем её
file_name = 'tasks.txt'
my_tasks = read_tasks(file_name)
````

````python
for task in my_tasks:
    print(task)
````

````python
# Объявляем функцию read_tasks_upd
def read_tasks_upd(text_file):
    # Создаем пустой список, где будут храниться преобразованные строки
    tasks_list = []
    # Открываем файл
    with open(file=text_file, encoding='utf-8') as f:
        # Считываем файл построчно
        for line in f:
            # Очищаем строку от знаков переноса
            strip_line = line.strip()
            # Добавляем очищенную строку в tasks_list
            tasks_list.append(strip_line)
    # Считаем кол-во элементов в списке tasks_list и сохраняем в переменную
    tasks_count = len(tasks_list)
    # В основном блоке `if/else проверяем, есть ли задачи, и выводим соответствующее сообщение
    if tasks_count == 0:
        print('Сегодня задач нет. Можно отдыхать.')
    else:
        # Во вложенной условной конструкции `if/else` проверяем число задач и выводим соответствующее сообщение
        if tasks_count >= 5:
            print('Сегодня много задач.')
        else:
            print('Сегодня задач немного.')
        # Поскольку задачи есть, выводим сообщение
        print('Задачи добавлены в список!')
    # Возвращаем список с задачами tasks_list
    return tasks_list
````

````python
# Создаем функцию-генератор с названием stream_gen
def stream_gen(course_name, n_streams):
    # Запускаем цикл, который создает последовательность от 0 до значения, находящегося в переменной n_streams
    for n in range(n_streams):
        # Создаем переменную course_stream, в которую сохраняем название курса, знак '_', номер потока +1, чтобы избежать значения 0 и включить крайнюю правую границу
        course_stream = course_name + '_' + str(n+1)
        # Возвращаем значение course_stream и приостанавливаем выполнение функции
        yield course_stream
````

````python
# Объявляем функцию для подсчета скидки, аргумент discount по умолчанию равен 0
def apply_discount(prices_dict, discount=0):
    # Создаем пустой словарь для добавления обновленных данных
    prices_dict_upd = {}
    # Запускаем цикл по переданному в функцию словарю
    for key in prices_dict:
        # Сохраняем значение по ключу в переменную
        price = prices_dict[key]
        # Высчитываем цену со скидкой и сохраняем результат в переменную
        price_with_discount = int(price - (price * discount / 100))
        # Добавляем в словарь новые данные
        prices_dict_upd[key] = price_with_discount
    # Возвращаем словарь
    return prices_dict_upd
````

````python
# Объявляем функцию calculate_metric
def calculate_metric(data, metric='revenue'):

    # Проверяем, относится ли метрика, переданная в metric_name, к отслеживаемым метрикам
    if metric not in ['revenue', 'orders', 'paying_users', 'aov', 'aov_alt']:
        print('Неверно указана метрика.')
        return

    # Считаем метрики
    if metric == 'revenue':
        revenue = 0
        for i in data:
            revenue += sum(i)
        return revenue

    if metric == 'orders':
        orders = 0
        for i in data:
            orders += len(i)
        return orders

    if metric == 'paying_users':
        paying_users = len(data)
        return paying_users

    if metric == 'aov':
        revenue = 0
        orders = 0
        for i in data:
            revenue += sum(i)
            orders += len(i)
        aov = round(revenue / orders, 2)
        return aov

    if metric == 'aov_alt':
        user_aov_list = [sum(i) / len(i) for i in data]
        aov = round(sum(user_aov_list) / len(user_aov_list), 2)
        return aov
````

Обычно [UTM-ссылки](https://tilda.education/articles-how-to-use-utm-url) выглядят примерно так:

`https://example.com/test_page/`

Затем после знака `?` через знак `&` перечисляются параметры с присвоенными им значениями:

* `utm_source=google`

* `utm_medium=cpc`

* `utm_campaign=my_promo`

В результате, если по ссылке будет совершён переход, система аналитики поймёт, что он был с источника `google` c типом трафика `cpc` (оплата за клик по рекламе), а сама рекламная кампания называлась `my_promo`.

**Решение**

````python
# Объявляем функцию utm_builder
def utm_builder(utm_params):
    # Создаем список для пропущенных параметров
    missing_params = []
    # Запускаем цикл по словарю
    for param in utm_params:
        # Проверяем на наличие пропущенных параметров и добавляем, если находим
        if utm_params[param] == '':
            missing_params.append(param)
    # Проверяем, есть ли в списке значения
    if len(missing_params) > 0:
        # Если список не пустой, выводим сообщение о пропущенных параметрах и завершаем работу функции
        params_join = ', '.join(missing_params)
        print(f'Ошибка. Не переданы параметры: {params_join}.')
        return
    # Создаем новый словарь, куда будем складывать исправленные значения параметров
    utm_params_ = {}
    # Запускаем еще один цикл по словарю
    for param in utm_params:
        # Преобразуем значения и сохраняем их в словарь
        utm_params_[param] = utm_params[param].strip().lower()
    # Достаем из нового словаря значения параметров по ключам
    website = utm_params_['website']
    course = utm_params_['course']
    source = utm_params_['utm_source']
    medium = utm_params_['utm_medium']
    campaign = utm_params_['utm_campaign']
    # Собираем адрес страницы
    website_url = f'https://{website}/{course}/'
    # Собираем UTM-ссылку
    utm_string = website_url + f'?utm_source={source}&utm_medium={medium}&utm_campaign={campaign}'
    # Возвращаем UTM-ссылку и завершаем работу функции
    return utm_string
````
